```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import load_wine, load_breast_cancer
```

**PERCEPTRON CLASSIFICATION**

WINE DATASET

```python
wine = load_wine()
df_wine = pd.DataFrame(wine.data)
df_wine
```

```python
x = wine.data
y = wine.target

#Klasifikasi Perceptron Perbandingan 0.2
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)

ppn_wine = Perceptron(eta0=0.1, random_state=0)
ppn_wine.fit(x_train_std, y_train)
y_pred = ppn_wine.predict(x_test_std)
accuracy = accuracy_score(y_test, y_pred)
print('Akurasi Dengan Perbandingan 0.2 adalah {:.2f}'.format(accuracy))

#Klasifikasi Perceptron Perbandingan 0.3
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)

ppn_wine = Perceptron(eta0=0.1, random_state=0)
ppn_wine.fit(x_train_std, y_train)
y_pred = ppn_wine.predict(x_test_std)
accuracy = accuracy_score(y_test, y_pred)
print('Akurasi Dengan Perbandingan 0.3 adalah {:.2f}'.format(accuracy))

#Klasifikasi Perceptron Perbandingan 0.4
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)

sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)

ppn_wine = Perceptron(eta0=0.1, random_state=0)
ppn_wine.fit(x_train_std, y_train)
y_pred = ppn_wine.predict(x_test_std)
accuracy = accuracy_score(y_test, y_pred)
print('Akurasi Dengan Perbandingan 0.4 adalah {:.2f}'.format(accuracy))
```

**BAGGING CLASSIFICATION**

WINE DATASET

```python
x = wine.data
y = wine.target

#Klasifikasi Bagging Dengan Perbandingan 100
ppn2_wine = Perceptron(eta0=0.1, random_state=0)
bgc = BaggingClassifier(estimator=ppn2_wine, base_estimators=100, n_jobs=-1)
bgc.fit(x_train_std, y_train)
y_pred = bgc.predict(x_test_std)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of Bagging Classifier Dengan Perbandingan 100 adalah {:.3f}'.format(accuracy))

#Klasifikasi Bagging Dengan Perbandingan 200
ppn2_wine = Perceptron(eta0=0.1, random_state=0)
bgc = BaggingClassifier(estimator=ppn2_wine, base_estimators=200, n_jobs=-1)
bgc.fit(x_train_std, y_train)
y_pred = bgc.predict(x_test_std)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of Bagging Classifier Dengan Perbandingan 200 adalah {:.3f}'.format(accuracy))

#Klasifikasi Bagging Dengan Perbandingan 300
ppn2_wine = Perceptron(eta0=0.1, random_state=0)
bgc = BaggingClassifier(estimator=ppn2_wine, base_estimators=300, n_jobs=-1)
bgc.fit(x_train_std, y_train)
y_pred = bgc.predict(x_test_std)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of Bagging Classifier Dengan Perbandingan 300 adalah {:.3f}'.format(accuracy))
```
